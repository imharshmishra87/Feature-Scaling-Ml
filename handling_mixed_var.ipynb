```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic.csv')
df.head()
```

# Dealing With 'Number' Column

The number column consist of mixed values stacked on top of each other

```python
df['numerical_column']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
df['numerical_column']
```

```python
# Addition wy to obtain the categorical column
df['list']=[df['number'][i] if df['numerical_column'].isna() [i] else np.nan for i in range(len(df['number']))]
df.drop(['list'],axis=1)
```

```python
df['categorical_column']=np.where(df['numerical_column'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

# Dealing With Cabin and Ticket column

The cabin columnn consists of values like C85

```python
df['cabin_numerical']=df['Cabin'].str.extract(r'(\d+)')
df['cabin_categorical']=df['Cabin'].str[0]
```

```python
df['Ticket_numerical']=df['Ticket'].apply(lambda s : s.split()[-1])
df['Ticket_numerical']=pd.to_numeric(df['Ticket_numerical'],errors='coerce',downcast='integer')
```

```python
df['Ticket_categorical']=df['Ticket'].apply(lambda s : s.split()[0])
df['Ticket_categorical']=np.where(df['Ticket_categorical'].str.isdigit(),np.nan,df['Ticket_categorical'])
```
