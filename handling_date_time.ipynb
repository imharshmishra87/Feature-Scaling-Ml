```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/gold_price_data.csv', usecols=['date'])
df.info()
```

# Converting The Date Datatype Into Dateformat from Object

```python
df['date']=pd.to_datetime(df['date'])
df.info()
```

## Extracting Year

```python
df['year']=df['date'].dt.year
df.sample(5)
```

## Extracting Month

```python
df['month']=df['date'].dt.month
df.sample(5)
```

## Extracting Month Name

```python
df['month_name']=df['date'].dt.month_name()
df.sample(5)
```

## Extracting Day

```python
# Specify the day of the month
df['day']=df['date'].dt.day
df.sample(5)
```

## Extracting The Day of The Week

```python
# Specify The Day of The Week
df['week_days']=df['date'].dt.dayofweek
df.head(10)
```

## Extracting the name of the day

```python
# Specifying the name of the day
df['day_name']=df['date'].dt.day_name()
df.sample(5)
```

## To check for weekends

```python
# To check for weekends
df['weekends']=np.where(df['day_name'].isin(['Friday']),1,0)
```

```python
df.head(10)
```

## Extracting Which week of th year it is

```python
# Extracting Which week of th year it is
df['which_week']=df['date'].dt.isocalendar().week
df.head()
```

## Extracting The Quarter

```python
df['Quarter']=df['date'].dt.quarter
df.head()
```

## Exracting The Semester

```python
df['Semester']=np.where(df['Quarter'].isin([1,2]),1,2)
df.head()
```

## Extracting Time-Elapsed Between Dates

```python
import datetime
today=datetime.datetime.today()
today
```

```python
today-df['date']
```

```python
(today-df['date']).dt.days
```

```python
np.round((today-df['date'])/np.timedelta64(1,'m'), 0)
```

# Dealing With Time

```python
time=pd.read_csv("/content/synthetic_datetime_dataset.csv",usecols=['Event_Timestamp'])
```

```python
time.info()
```

## Changing The DataType

```python
time['Event_Timestamp']=pd.to_datetime(time['Event_Timestamp'])
time.info()
```

## Extracting The Time

```python
time['time']=time['Event_Timestamp'].dt.time
time.head()
```

## Extracting The Minute

```python
time['minutes']=time['Event_Timestamp'].dt.minute
time.head()
```

## Extracting The Hour

```python
time['hour']=time['Event_Timestamp'].dt.hour
time.head()
```

# Calculating Time Difference

```python
(today-time['Event_Timestamp'])
```


## Calculating Time Difference in minutes

```python
np.round((today-time['Event_Timestamp'])/np.timedelta64(1,'m'))
```

## Calculating Time Difference in seconds

```python
np.round((today-time['Event_Timestamp'])/np.timedelta64(1,'s'))
```

## Calculating Time Difference in hours

```python
np.round((today-time['Event_Timestamp'])/np.timedelta64(1,'h'))
```
